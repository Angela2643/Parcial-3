```python
# Paradigma Imperativo: Sistema de Gestión de Estudiantes

# Lista principal para almacenar estudiantes (matriz de diccionarios)
estudiantes = []

# Función para agregar un nuevo estudiante
def agregar_estudiante():
    nombre = input("Ingrese el nombre del estudiante: ")
    nuevo_estudiante = {
        "nombre": nombre,
        "calificaciones": [],
        "promedio": 0.0
    }
    estudiantes.append(nuevo_estudiante)
    print(f"Estudiante {nombre} agregado exitosamente!\n")

# Función para agregar calificaciones a un estudiante
def agregar_calificaciones():
    if len(estudiantes) == 0:
        print("No hay estudiantes registrados!\n")
        return

    print("Lista de estudiantes:")
    for i, estudiante in enumerate(estudiantes):
        print(f"{i+1}. {estudiante['nombre']}")

    try:
        indice = int(input("\nSeleccione el número del estudiante: ")) - 1
        if 0 <= indice < len(estudiantes):
            calificacion = float(input("Ingrese la calificación: "))
            estudiantes[indice]["calificaciones"].append(calificacion)
            print("Calificación agregada exitosamente!\n")
        else:
            print("Número de estudiante inválido!\n")
    except ValueError:
        print("Entrada inválida, debe ser un número!\n")

# Función para calcular promedios
def calcular_promedios():
    for estudiante in estudiantes:
        if len(estudiante["calificaciones"]) > 0:
            suma = sum(estudiante["calificaciones"])
            promedio = suma / len(estudiante["calificaciones"])
            estudiante["promedio"] = round(promedio, 2)
        else:
            estudiante["promedio"] = 0.0
    print("Promedios actualizados!\n")

# Función para mostrar todos los estudiantes
def mostrar_estudiantes():
    if len(estudiantes) == 0:
        print("No hay estudiantes registrados!\n")
        return

    print("\n=== LISTA DE ESTUDIANTES ===")
    for estudiante in estudiantes:
        print(f"Nombre: {estudiante['nombre']}")
        print(f"Calificaciones: {estudiante['calificaciones']}")
        print(f"Promedio: {estudiante['promedio']}\n")

# Función principal con menú interactivo
def menu():
    while True:
        print("=== SISTEMA DE GESTIÓN ESCOLAR ===")
        print("1. Agregar estudiante")
        print("2. Agregar calificación")
        print("3. Calcular promedios")
        print("4. Mostrar todos los estudiantes")
        print("5. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":
            agregar_estudiante()
        elif opcion == "2":
            agregar_calificaciones()
        elif opcion == "3":
            calcular_promedios()
        elif opcion == "4":
            mostrar_estudiantes()
        elif opcion == "5":
            print("¡Hasta luego!")
            break
        else:
            print("Opción inválida, intente nuevamente\n")

# Ejecutar el programa
menu()
```
